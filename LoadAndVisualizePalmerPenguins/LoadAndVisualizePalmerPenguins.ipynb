```python
# Before we do anything, we first import the libraries we will use.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

We first need to download data from: https://allisonhorst.github.io/palmerpenguins/articles/download.html

There is one dataset for each penguin species. They were downloaded as "table_219.csv", "table_220.csv" and "table_221.csv". Before running the following cells, make sure the files are located in the same directory where this notebook is saved. 

It is best practice to not include ["absolute paths"](https://www.redhat.com/sysadmin/linux-path-absolute-relative#:~:text=An%20absolute%20path%20makes%20no,up%20at%20the%20target%20location.) on your Notebooks, since the code would then only be able to run in your personal computer. Instead use "relative paths". 


```python
# Load 3 datasets

penguins1=pd.read_csv("./table_219.csv")
penguins2=pd.read_csv("./table_220.csv")
penguins3=pd.read_csv("./table_221.csv")
```

We used the read_csv function because we were working with files with comma-separated values. 

However, Pandas can load data from multiple sources. You can find a list of functions here: https://pandas.pydata.org/pandas-docs/stable/reference/io.html

Some that may be interesting are `pd.read_excel()` or `pd.read_clipboard()`. 


```python
# OPTIONAL
# We can take a look at each dataset by just typing their name and running this cell
# You can take a look at each dataset by modifying the line below. How would you modify it?

penguins3
```

Since we downloaded each dataset separately we need to use the function pd.concat to concatenate them. 

In this case it is very straightforward because the 3 datasets have the same format. 

```python
# Concatenate 3 datasets into 1 

penguins = pd.concat([penguins1,penguins2,penguins3],ignore_index=True)
```

```python
# Let's take a look at how the new dataset looks like. 

penguins
```

# Data Visualization

We will use the [seaborn library](https://seaborn.pydata.org/tutorial/introduction) to visualize the data. Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with the pandas data structures. 

Most of the plotting functions are structured around three modules: “relational”, “distributional”, and “categorical”.

<img src="https://seaborn.pydata.org/_images/function_overview_8_0.png" alt="seaborn modules" style="width:500px;"/>







We already loaded the penguins dataset into the pandas dataframe "penguins", so we can already start plotting! 

```python
# Run this cell to plot "Culmen Depth (mm)" vs "Culmen Length (mm)", for each of the species. 

sns.set_theme()

sns.relplot(
    data=penguins,
    x="Culmen Length (mm)", 
    y="Culmen Depth (mm)", 
    col="Species")
```


<img src="https://allisonhorst.github.io/palmerpenguins/reference/figures/culmen_depth.png" alt="penguins_bills" style="width:500px;"/>


```python
# We can now plot all the data into a single plot, and color each species in a different color

sns.relplot(
    data=penguins,
    x="Flipper Length (mm)", 
    y="Body Mass (g)", 
    hue="Species")
```

```python
sns.displot(data=penguins, x="Flipper Length (mm)", hue="Species", col="Species")
```

```python
sns.catplot(data=penguins, x="Flipper Length (mm)", y="Species",kind="box")
```

## How do we save to disk? 

So far we have only displayed the plots inside the notebook. However, we need to learn how to save the plots to disk. We can do this using the function `savefig` from the `matplotlib` library. When using `Seaborn` in `Jupyter Notebook`, you need to run the plotting and saving functions inside the same cell, otherwise an empty file is created. 

You can specify the name and path of the file. In the example below, we save the figure in the local directory (the same one where you opene this notebook). 

```python
sns.catplot(data=penguins, x="Flipper Length (mm)", y="Species",kind="box")
plt.savefig("./seaborn_plot.jpg", dpi=300)
```

# Multivariate Views on Complex Datasets

Some seaborn functions combine multiple kinds of plots to quickly give informative summaries of a dataset.

```python
# jointplot() shows the relationship between two variables, 
# as well as plotting each variable's marginal distribution

sns.jointplot(data=penguins, x="Flipper Length (mm)", y="Culmen Length (mm)", hue="Species")
```

```python
# pairplot(), plots all the variable combinations and the 
# marginal distribution for each variable

sns.pairplot(data=penguins, hue="Species")
```
